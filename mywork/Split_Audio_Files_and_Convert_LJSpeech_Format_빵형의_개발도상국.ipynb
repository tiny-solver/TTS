```python
from pytube import YouTube

# 아델 목소리
yt = YouTube('https://www.youtube.com/watch?v=dpbb81nUqUg')

yt.streams.filter(only_audio=True).first().download(
    output_path='.', filename='input.mp3')
```

## Transcribe

```python
import whisper

model = whisper.load_model("small")
```

30초 이상 오디오 데이터 처리 방법

https://github.com/openai/whisper/discussions/136

```python
result = model.transcribe("input.mp3")

result["text"][:300]
```

```python
result["segments"][:3]
```

```python
len(result["text"])
```

```python
for r in result['segments']:
    print(f'[{r["start"]:.2f} --> {r["end"]:.2f}] {r["text"]}')
```

```python
!mkdir wavs
```

```python
for i, r in enumerate(result['segments']):
    !ffmpeg -y -i input.mp3 -ss {r["start"]} -to {r["end"]} -hide_banner -loglevel error wavs/audio{i+1}.wav
```

```python
# result["text"][303]
for i, r in enumerate(result['segments']):
  if i == 29:
    print(r["start"], r["end"])
```

```python
from IPython.display import Audio

Audio("wavs/audio30.wav")
```

```python
with open("metadata.txt", "w", encoding="utf-8") as f:
    for i, r in enumerate(result['segments']):
        f.write(f"audio{i+1}|{r['text'].strip()}|{r['text'].strip()}\n")
```
